# 📊 Student Performance Analysis


## 📌 Objective

The objective of this project is to analyze student performance data and understand how different factors affect the final grade (G3).

## 📂 Dataset Description

The dataset contains the following features:
- School
- Gender (sex)
- Age
- Study time
- Family size and background
- Health condition
- Absences
- Grades (G1, G2, G3)

## 🛠️ Libraries Used

- Pandas → Data manipulation
- Matplotlib → Data visualization

## 📥 Data Loading

The dataset is loaded using pandas and stored in a DataFrame for further analysis.

## 🧹 Data Cleaning

- Checked for missing values
- Removed duplicate rows
- Verified data types of each column

## 🔍 Basic Analysis

- Calculated average final grade (G3)
- Counted number of students scoring above 15

## 📊 Visualization

A histogram is plotted to show the distribution of final grades (G3).

## 📈 Correlation Analysis

The relationship between study time and final grade (G3) is analyzed using correlation.

## 👥 Group Analysis

The average grade is calculated based on gender using groupby function.

## 📌 Results

- Average G3 ≈ 10.41
- Students scoring above 15 ≈ 40
- Weak positive correlation between study time and grades

## ✅ Conclusion

Student performance depends on multiple factors. Study time has some effect, but other variables also influence final grades.
